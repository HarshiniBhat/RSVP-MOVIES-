```python
    # Group the movies by individual genres

df_by_g1 = df_by_genre.groupby( by = 'genre_1').sum()
df_by_g1
```

```python
df_by_g2 = df_by_genre.groupby( by = 'genre_2').sum()
df_by_g2
```

```python
df_by_g3 = df_by_genre.groupby( by = 'genre_3').sum()
```

```python
# Add the grouped data frames and store it in a new data frame
df_by_g1.add(df_by_g2, fill_value = 0).add(df_by_g3, fill_value = 0)
```

```python
# creating a new data frame 
df_add = df_by_g1.add(df_by_g2, fill_value = 0).add(df_by_g3, fill_value = 0)
df_add.head()
```

```python
# Extract genres with atleast 10 occurences

df_add[df_add.cnt >=10]
```

```python
# creating genre_top10 data frame 
genre_top10 = df_add[df_add.cnt >=10]
genre_top10 
```

```python
# Take the mean for every column by dividing with cnt 

genre_top10_mean  = genre_top10.iloc[:,0:-1].div(genre_top10.cnt,axis = 0)
genre_top10_mean
```

```python
genre_top10 = pd.concat([genre_top10_mean, genre_top10.cnt],axis =1 )
genre_top10.head()
```

```python
# Rounding off the columns of Votes to two decimals

genre_top10.loc[:,'VotesM': 'VotesnUS'].round(2)
```

```python
# Converting CVotes to int type

genre_top10.loc[:,'CVotes10':'CVotesUS'].astype('int')
```

```python
genre_top10 = pd.concat([genre_top10.loc[:,'CVotes10':'CVotesUS'].astype('int'),genre_top10.loc[:,'VotesM':'VotesnUS'],genre_top10.cnt], axis =1)
genre_top10.head()
```
